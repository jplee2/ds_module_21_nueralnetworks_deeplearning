## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN", "NAME"])
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()
unique_values
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_counts = application_df['APPLICATION_TYPE'].value_counts()
application_counts
```

```python
# Define a threshold for grouping
threshold = 200  # Adjust this based on your dataset

# Get the value counts for APPLICATION_TYPE
application_counts = application_df['APPLICATION_TYPE'].value_counts()

# Create a list of application types to be replaced
application_types_to_replace = application_counts[application_counts < threshold].index.tolist()
#Replace the application types in the DataFrame:

for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")
#Check to make sure the replacement was successful:

application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts.loc[classification_counts > 1]
```

```python
# Define a threshold for grouping
threshold = 1500  # You can adjust this based on your dataset

# Create a list of classifications to be replaced
classifications_to_replace = classification_counts[classification_counts < threshold].index.tolist()
#Replace those classifications with "Other":

for classification in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(classification, "Other")

#Finally, check to make sure the replacement was successful:
print(application_df['CLASSIFICATION'].value_counts())
```

```python
# Optional Extra Work:
# bucket ORGANIZATION, USE_CASE, AFFILIATION (even though less than 10 categories, they have small sample sizes)
# & drop SPECIAL_CONSIDERATIONS, STATUS (due to category value_count discrepencies)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummy_df = pd.get_dummies(application_df, dtype=int)
dummy_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = dummy_df.drop(columns=["IS_SUCCESSFUL"])
y = dummy_df.IS_SUCCESSFUL

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=1, stratify=y)
print(X_train.shape, X_test.shape)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train["ASK_AMT"].values.reshape(-1,1))

# Scale the data
X_train_scaled = X_train.copy()
X_train_scaled["ASK_AMT"] = X_scaler.transform(X_train["ASK_AMT"].values.reshape(-1, 1))
X_train_scaled.head()
```

```python
X_test_scaled = X_test.copy()
X_test_scaled["ASK_AMT"] = X_scaler.transform(X_test["ASK_AMT"].values.reshape(-1, 1))
```

## Compile, Train and Evaluate the Model

```python
# Create the Keras Sequential model
nn_model = tf.keras.models.Sequential()

# Add our first Dense layer, including the input layer
nn_model.add(tf.keras.layers.Dense(units=7, activation="relu", input_dim=len(X.columns)))

# Add in a second layer
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu"))

# Add in a third layer
nn_model.add(tf.keras.layers.Dense(units=3, activation="relu"))

# Add the output layer that uses a probability activation function
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the Sequential model
nn_model.summary()
```

```python
# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model to the training data
fit_model = nn_model.fit(X_train, y_train, epochs=500)
```

```python
# Evaluate the model using the test data
test_loss, test_accuracy = nn_model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model.save("my_model.h5")
```
